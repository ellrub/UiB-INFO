```python
class Tekstanalyse:

    def normaliser_tekst(self, filnavn):
        "Fjerner spesialtegn fra self.tekst og konverterer til små bokstaver."
        with open(filnavn) as file:
            tekst = file.read()
            normal_tekst = "".join(character for character in tekst if character.isalnum() or character.isspace())
            return normal_tekst.lower()

    def til_avsnitt(self, normal_tekst):
        "Deler self.tekst opp i en liste av avsnitt som lagres i self.avsnitt ."
        avsnitt = normal_tekst.split("\n\n")
        return avsnitt

    def lag_ordliste(self, avsnitt):
        "Lager en liste av ord som forekommer i avsnittet."
        ordliste = []
        for ord in avsnitt:
            ordliste.append(ord.split())
        return ordliste

    def tell_ordforekomster(self, avsnitt):
        "Lager en liste over antall forekomster av ordene i ordliste i avsnittet."
        antall_ord = []
        for liste in avsnitt:
            ord_dict = {}
            for ord in liste:
                if ord in ord_dict:
                    ord_dict[ord] += 1
                else:
                    ord_dict[ord] = 1
            antall_ord.append(ord_dict)
        return antall_ord

    def skriv_ut(self, filnavn):
        "Skriver ut analyseresultatene for hvert avsnitt på skjermen."
        normal_tekst = self.normaliser_tekst(filnavn)
        avsnitt = self.til_avsnitt(normal_tekst)
        ordliste = self.lag_ordliste(avsnitt)
        antall_ord = self.tell_ordforekomster(ordliste)
        for i in range(len(avsnitt)):
            print(f"Avsnit {i + 1}")
            print(f"Avsnittekst: {avsnitt[i]}")
            print(f"Ordtelling / Ordliste: {antall_ord[i]}\n")

    def lagre_til_fil(self, filnavn):
        "Lagrer analyseresultatene for hvert avsnitt i en fil."
        with open("analyseresultat.txt", "w") as file:
            normal_tekst = self.normaliser_tekst(filnavn)
            avsnitt = self.til_avsnitt(normal_tekst)
            ordliste = self.lag_ordliste(avsnitt)
            antall_ord = self.tell_ordforekomster(ordliste)
            for i in range(len(avsnitt)):
                file.write(f"Avsnit {i + 1}\n")
                file.write(f"Avsnittekst: {avsnitt[i]}\n")
                file.write(f"Ordtelling / Ordliste: {antall_ord[i]}\n\n")

tekstanalyse = Tekstanalyse()
tekstanalyse.lagre_til_fil("eksempeltekst.txt")
# tekstanalyse.skriv_ut("eksempeltekst.txt")
```
