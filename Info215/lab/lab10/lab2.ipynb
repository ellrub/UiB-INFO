```python
import networkx as nx
```

```python
G = nx.Graph()

G.add_edge('A', 'B')
G.add_edge('A', 'C')
G.add_edge('A', 'E')
G.add_edge('C', 'E')
G.add_edge('C', 'D')
```

```python
pos = {'A': (0, 0), 'B': (0, -1), 'C': (1, -1), 'D': (2, -1), 'E': (1, 0)}
nx.draw(G, pos, with_labels=True)
```
