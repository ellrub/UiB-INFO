```python
import networkx as nx
```

```python
G = nx.Graph()

# Node 0
G.add_edge('0', '1')
G.add_edge('0', '2')
G.add_edge('0', '3')
G.add_edge('0', '4')
G.add_edge('1', '3')
G.add_edge('2', '3')
G.add_edge('2', '4')
G.add_edge('2', '5')
G.add_edge('3', '5')
```

```python
pos = nx.spring_layout(G, seed=6)
nx.draw(G, pos, with_labels=True)
```
