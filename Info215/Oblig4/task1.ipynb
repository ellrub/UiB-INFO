```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
G = nx.karate_club_graph()

degrees = G.degree()
nodes = G.nodes()
```

```python
# Task 1.1
print(G.nodes(data = True))
print(G.edges(data = True))
```

```python
# Task 1.2
colors = {
    "Mr. Hi": "#ffb233",
    "Officer": "#4d95f1"
}
n_color = [colors[G.nodes[n]['club']] for n in G.nodes]
n_size = np.asarray([degrees[n] * 50 for n in nodes])



pos = nx.spring_layout(G, seed = 6)

options = {
    "node_color": n_color,
    "node_size": n_size,
    "edge_color": "grey",
    "linewidths": 0,
    "width": 0.4,
    "with_labels" : True,
}

nx.draw(G, pos, **options)
```

```python
# Task 1.3
print(nx.dijkstra_path(G, 24, 16))
```

```python
# Task 1.4
djikstra_path_nodes = nx.dijkstra_path(G, 24, 16)
colors = {
    "Mr. Hi": "#ffb233",
    "Officer": "#4d95f1"
}
n_color = [colors[G.nodes[n]['club']] for n in G.nodes]
n_size = np.asarray([degrees[n] * 50 for n in nodes])

for n in G.nodes:
    if n in djikstra_path_nodes:
        n_color.pop(n)
        n_color.insert(n, "#f63a3a")

pos = nx.spring_layout(G, seed = 6)

options = {
    "node_color": n_color,
    "node_size": n_size,
    "edge_color": "grey",
    "linewidths": 0,
    "width": 0.4,
    "with_labels" : True,
}

nx.draw(G, pos, **options)
```
