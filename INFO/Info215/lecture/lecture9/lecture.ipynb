```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
G.add_edge('v1', 'v2')
G.add_edge('v2', 'v3')
G.add_edge('v2', 'v4')
G.add_edge('v4', 'v5')
options = {
    'node_color': [20, 12, 12, 8, 2], # Adjust numbers up or down to change colour strength on nodes
    'node_size': 700,
    'cmap': plt.cm.Oranges,
    'edge_color': 'grey',
    'linewidths': 0,
    'width': 0.4,
    'with_labels': True,
}
pos = {'v1': (0, 0), 'v2': (3, 1), 'v3': (3, -1), 'v4': (7, 1), 'v5': (7, -1)}
nx.draw(G, pos, **options)
```

```python
nx.write_graphml(G, 'nutrients.graphml')

# or

# with open('nutrients.graphml', 'wb')as ofile:
#     nx.write_graphml(G, ofile)
```
